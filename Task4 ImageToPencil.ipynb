#Image to pencil sketch using Python
The image read in RBG format, then converted into a grayscale image. The grayscale image is then inverted to obtain a negative image ehich is done in order to enhance details. Finally, the pencil sketch is created by mixing the grayscale image with the inverted blurry image

```python
import cv2
```

```python
from matplotlib import pyplot as plt
```

```python
img=cv2.imread('/content/ItoP.jpg')
img.shape
```

#Color

```python
color=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,10))
plt.imshow(color)
plt.title('Taj Mahal --- Color')
plt.axis('off')
plt.show()
```

#GRAY SCALE

```python
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(10,10))
plt.imshow(gray,cmap='gray')
plt.title('Taj Mahal -- Gray')
plt.axis('off')
plt.show()
```

#INVERTED IMAGE

```python
inv=cv2.bitwise_not(gray)
plt.figure(figsize=(10,10))
plt.imshow(inv,cmap='gray')
plt.title('Taj Mahal -- Inverted Image')
plt.axis('off')
plt.show()
```

#BLURRED INV IMAGE

```python
blurred=cv2.GaussianBlur(inv,(21,21),0)
plt.figure(figsize=(10,10))
plt.imshow(blurred,cmap='gray')
plt.title('Taj Mahal -- Blurred')
plt.axis('off')
plt.show()
```

#PENCIL SKETCH

```python
pencilSketch=275-blurred
```

```python
sketch=cv2.divide(gray,inv,scale=275)
inv=cv2.bitwise_not(gray)
plt.figure(figsize=(12,12))
plt.imshow(sketch,cmap='gray')
plt.title('Taj Mahal -- PencilSketch')
plt.axis('off')
plt.show()
```
